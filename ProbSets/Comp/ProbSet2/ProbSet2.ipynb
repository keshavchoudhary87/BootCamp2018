```python
import numpy as np
from matplotlib import pyplot as plt

def func1(n):
    B = np.var(np.mean(np.random.normal(size=(n,n)),axis=1))
    return B

def func2():
    A = np.arange(100,1100,100)
    X = np.empty(len(A))
    for i in range(len(X)):
        X[i] = func1(A[i])
    return plt.plot(A,X)
func2()
    
    

```

```python
def func():
    x = np.linspace(-2*np.pi, 2*np.pi, 100)
    y = x.copy()
    X,Y = np.meshgrid(x,y)
    gxy = ((np.sin(X)*np.sin(Y))/(X*Y))
    plt.subplot(121)
    plt.pcolormesh(X,Y,gxy, cmap='viridis')
    plt.colorbar()
    plt.subplot(122)
    plt.contour(X,Y,gxy,10, cmap='coolwarm')
    plt.colorbar()
    plt.xlim(-2*np.pi, 2*np.pi)
    plt.ylim(-2*np.pi, 2*np.pi)
    plt.show
    
    
func()
```

# Data Visualization

Problem 1

```python
import numpy as np
from scipy import stats
from matplotlib import pyplot as plt

data =np.load('anscombe.npy')

def func():
    for i in range(4):
        plt.subplot(2,2,(i+1))
        plt.plot(data[:, 2*(i)], data[:, 2*(i)+1], 'o')
        slope, intercept, r_value, p_value, std_err = stats.linregress(data[:, 2*i], data[:, 2*i + 1])
        print('Mean, variance of x in regression number '+str(i+1)+' is: ', np.mean(data[:, 2*i]), np.var(data[:,2*i]))
        print('Mean, variance of y in regression number '+str(i+1)+' is: ', np.mean(data[:, 2*i+1]), np.var(data[:, 2*i+1]))
        print('Slope, intercept in regression number '+str(i+1)+' is: ', slope, intercept)
        plt.plot(data[:, 2*i], intercept + slope*data[:,2*i])
        plt.xlim(0,20)
        plt.tight_layout()                                                              
    return plt.show()

    
func()
    
```

Problem 2

```python
from scipy import special 
x = np.linspace(0,1,100)

def bernstein(n,v):
    A = special.binom(n,v)*(x**v)*(1-x)**(n-v)
    return A

fig,ax = plt.subplots(4,4)

for i in range(4):
    for j in range(i+1):
        ax[i,j].plot(x, bernstein(i,j))
        ax[i,j].set_title('n= '+str(i)+', v= '+str(j))
plt.tight_layout() 
plt.show()


    
```

```python
data = np.load('MLB.npy')
height = data[:,0]
weight = data[:,1]
age = data[:,2]

fig,ax = plt.subplots(1,3)
ax[0].plot(height, weight, '.')
slope, intercept, r_value, p_value, std_err = stats.linregress(height, weight)
ax[0].plot(height, slope*height + intercept)
ax[0].set_xlabel('height')
ax[0].set_ylabel('weight')

ax[1].plot(age, weight,'.')
slope, intercept, r_value, p_value, std_err = stats.linregress(age, weight)
ax[1].plot(age, slope*age + intercept)
ax[1].set_xlabel('age')
ax[1].set_ylabel('weight')

ax[2].plot(age, height,'.')
slope, intercept, r_value, p_value, std_err = stats.linregress(age, height)
ax[2].plot(age, slope*age + intercept)
ax[2].set_xlabel('age')
ax[2].set_ylabel('height')

plt.tight_layout()
plt.show()





```
