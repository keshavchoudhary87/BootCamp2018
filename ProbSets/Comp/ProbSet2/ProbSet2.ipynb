Problem 1

```python
import numpy as np
from scipy import stats
from matplotlib import pyplot as plt

data =np.load('anscombe.npy')

def func():
    for i in range(4):
        plt.subplot(2,2,(i+1))
        plt.plot(data[:, 2*(i)], data[:, 2*(i)+1], 'o')
        slope, intercept, r_value, p_value, std_err = stats.linregress(data[:, 2*i], data[:, 2*i + 1])
        print('Mean, variance of x in regression number '+str(i+1)+' is: ', np.mean(data[:, 2*i]), np.var(data[:,2*i]))
        print('Mean, variance of y in regression number '+str(i+1)+' is: ', np.mean(data[:, 2*i+1]), np.var(data[:, 2*i+1]))
        print('Slope, intercept in regression number '+str(i+1)+' is: ', slope, intercept)
        plt.plot(data[:, 2*i], intercept + slope*data[:,2*i])
        plt.xlim(0,20)
        plt.tight_layout()                                                              
    return plt.show()

    
func()
    
```

```python
from scipy import special 
x = np.linspace(0,1,100)

def bernstein(n,v):
    A = special.binom(n,v)*(x**v)*(1-x)**(n-v)
    return A

fig,ax = plt.subplots(4,4)

for i in range(4):
    for j in range(i+1):
        ax[i,j].plot(x, bernstein(i,j))
        ax[i,j].set_title('n= '+str(i)+', v= '+str(j))
plt.tight_layout() 
plt.show()


    
```

```python
def rosenbrock():
    x = np.linspace(-2,2,100)
    y = np.linspace(-2,2,100)
    X,Y = np.meshgrid(x,y)
    Z = ((1-X)**2) + 100*(Y - X**2)**2
    fig,ax = plt.subplots(1,2)
    plt.axis('equal')
    ax[0].pcolormesh(X,Y,Z, cmap='viridis')
    ax[1].contour(X,Y,Z, [-2,-1.5,-1,-0.5, -0.25,0.25, 0.5, 1, 1.5, 2,3,4,10,20,30,40])
    return plt.show()

rosenbrock()
```
