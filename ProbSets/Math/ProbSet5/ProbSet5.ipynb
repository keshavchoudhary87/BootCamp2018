```python
import numpy as np
import matplotlib.pyplot as plt
```

Problem 8.1

```python
x = np.linspace(0,6.5,200)
y1 = (2/3)*x + (4/3)
y2 = (1/6)*x -(1/6)
y3 = 6 - x
y6 = np.maximum(y3, y1)
y=[]
for i in range(3):
    y.append((5/4)*x - i)

plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)
for i in range(3):
    plt.plot(x, y[i], label='objective function')
plt.fill_between(x, y1, y6, label='feasible set')
plt.xlim(-0,4)
plt.ylim(0,7)
plt.title('Maximization problem')
plt.legend()
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
print('The optimal value of the problem is at (x=2.8, y=3.2) where the objective function takes the value 1.2')
```

Problem 8.2 (i)

```python
x = np.linspace(0,100,1000)
y1 = 5 - x/3
y2 = 6 - 2*(x/3)
y3 = x-4
y4 = np.minimum(y1,y2)
y5 = np.minimum(y3,y4)



plt.plot(x, y1, label= '$y_1= 5-x/3$')
plt.plot(x, y2, label= '$y_2= 6 - 2x/3$')
plt.plot(x, y3, label= '$y_3= x-4$')
plt.fill_between(x, 0, y4, where= x<4, color='b')
plt.fill_between(x, y3, y4, where= x<6, label='feasible set')
plt.plot(x, 20-3*x, label='objective function')
plt.legend()
plt.xlim(0,17.5)
plt.ylim(0,7)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()

print('The objective function is maximised at (6,2)')
print('The value of the objective function is 27')
```
